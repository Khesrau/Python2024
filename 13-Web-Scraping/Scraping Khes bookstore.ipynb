```python
titles=[]
prices=[]
```

```python
import time
```

```python
def book_scrapping(pages=14,max_price=50):
    
    titles=[]
    prices=[]
    base_url=input("Please provide the base url. Replace that page indicator with '{}'")
    
    for page in range(pages):
        res=requests.get(base_url.format(page))
        soup=bs4.BeautifulSoup(markup=res.text,features="lxml")
        books=soup.select(selector=".col-md-12.col-xs-12.mt-product2.large.bg-grey.p5")
        for book in books:
            if float(book.select("b")[1].text[::-1][3:][::-1])<max_price:

                titles.append(book.select(".sstitle")[0].text)
                prices.append(float(book.select("b")[1].text[::-1][3:][::-1]))
    
    
    import pandas as pd
    df=pd.DataFrame({"Titles":titles,"Prices":prices})
    
    return df
    # create pandas DataFrame
    # Time the code
```
